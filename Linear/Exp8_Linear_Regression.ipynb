```python
# Coding the Algorithm

data = [
    [20,0.18],
    [60,0.37],
    [100,0.35],
    [140,0.78],
    [180,0.56],
    [220,0.75],
    [260,1.18],
    [300,1.36],
    [340,1.17],
    [380,1.65]
]

x = [i[0] for i in data]
y = [i[1] for i in data]
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
import math

def mean(n):
    return sum(n) / float(len(n))

def covariance(x, mx, y, my):
    covar = 0.0
    for i in range(len(x)):
        covar += (x[i] - mx) * (y[i] - my)
    return covar

def variance(values, mean):
    return sum([(x-mean)**2 for x in values])
```

```python
x_mean, y_mean = mean(x), mean(y)
w1 = covariance(x, x_mean, y, y_mean) / variance(x, x_mean)
w0 = y_mean - w1 * x_mean
print(w0)
print(w1)
```

```python
import numpy as np
x = np.array(x)
plt.scatter(x,y)
plt.plot(x, w0 + w1*x, linestyle='solid')
plt.show()
```

```python
def rmse(actual, predicted):
    sum_error = 0.0
    for i in range(len(actual)):
        prediction_error = predicted[i] - actual[i]
        sum_error += (prediction_error ** 2)
    mean_error = sum_error / float(len(actual))
    return (mean_error)**0.5

yPred = [w0 + q*w1 for q in x]
print(rmse(y,yPred))
```

```python
print("Enter the x to classify")
test = [int(i) for i in input().split()]
yNew = w0 + w1 * test[0]
print(yNew)
```

```python
# Using in-built libraries

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Data
data = np.array([
    [20, 0.18],
    [60, 0.37],
    [100, 0.35],
    [140, 0.78],
    [180, 0.56],
    [220, 0.75],
    [260, 1.18],
    [300, 1.36],
    [340, 1.17],
    [380, 1.65]
])

# Separate data into X (features) and y (target)
X = data[:, 0].reshape(-1, 1)
y = data[:, 1]
```

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Get the intercept and coefficient
w0 = model.intercept_
w1 = model.coef_[0]
```

```python
# Print intercept and coefficient
print("Intercept (w0):", w0)
print("Coefficient (w1):", w1)
```

```python
# Plot the data and the regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X), color='red', linestyle='solid')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Calculate RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y, model.predict(X)))
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
# Predict for new data point
print("Enter the x to classify:")
test_x = np.array([[int(input())]])
y_new = model.predict(test_x)
print("Predicted y for the new x:", y_new[0])
```
