```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.hist(df['Age'], bins = 10)

plt.title("Histogram")
plt.xlabel("Age")
plt.ylabel("Patients")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.scatter(df['Age'],df['MaxHR'])
plt.title("Scatter plot of MaxHR and Age")
plt.xlabel("Age")
plt.ylabel("Patients")
```

```python
# Map 0 to 'Male' and 1 to 'Female'
df['Sex'] = df['Sex'].map({0: 'Male', 1: 'Female'})

# Calculate the counts of males and females
sex_counts = df['Sex'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.title('Distribution of Males and Females')
plt.show()
```

```python
# Correlation Matrix

# Correlation coefficient b/w Age & RestBP
matrix = np.corrcoef(df['Age'], df['RestBP'])
print(matrix)
```

```python
# Replace missing value in the dataframe

# finding mean ca column
meanCa = np.mean(df.Ca)

# fill the missing values (na => not available values) with the mean value
df.Ca = df.Ca.fillna(meanCa)

df.isna().sum()
```

```python
# To remove missing values
df = df.dropna()

# Now we'll see count of missing value in each column
df.isna().sum()
```
