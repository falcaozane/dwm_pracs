```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 5) Describe the attributes find range, quartile, percentile
# percentile, box plot and outliers

df.describe()
# describe function gives overall summary of your data.
```

```python
# Calculate quartile as per RestBP column
print('Q1 => 25th Percentile of arr : ', np.percentile(df['RestBP'], 25))
print('Q2 => 50th Percentile of arr : ', np.percentile(df['RestBP'], 50))
print('Q3 => 75th Percentile of arr : ', np.percentile(df['RestBP'], 75))
```

```python
np.quantile(df['RestBP'], [0, 0.25, 0.5, 0.75, 1])
```

```python
# Boxplot
sns.boxplot(df)
sns.despine() #removes top & right border from the plot
```

```python
# Histogram
fig = plt.figure() #canvas
ax = fig.add_subplot(1,1,1) #1 subplot in 1 row & 1 column

# Variable
# Histogram of Age column, divided into 10 age intervals(10bars), to indicate frequency of that age group..
ax.hist(df['Age'], bins = 10)

# Label & Title
plt.title('Age Distribution')
plt.xlabel("Age")
plt.ylabel("Patients")

plt.show()
```

```python
# Scatter Plot

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Variable
# Scatter plot to show relation b/w Age & MaxHR
ax.scatter(df['Age'], df['MaxHR'])

# Labels & title
plt.title("MaxHR & Age Distribution")
plt.xlabel("Age")
plt.ylabel("MaxHR")

plt.show()
```

```python
# Labels for age groups
age_labels = ['30-39', '40-49', '50-59', '60+']
age_ranges = [(30, 39), (40, 49), (50, 59), (60, float('inf'))]

# Categorize ages into age groups
age_counts = [((df['Age'] >= start) & (df['Age'] <= end)).sum() for start, end in age_ranges]

plt.pie(age_counts, labels=age_labels, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Age Distribution')

# Display the pie chart
plt.show()
```

```python
# Frequency Table
# 4 chestpains categories for each sex.
test = df.groupby(['Sex', 'ChestPain'])
test.size() #gives count
```

```python
# Correlation Matrix

# Correlation coefficient b/w Age & RestBP
matrix = np.corrcoef(df['Age'], df['RestBP'])
print(matrix)
```

```python
# Identify Missing Values & Outliers then fill them with avg.

# Check for missing values in each column
# na => not available,
df.isna().sum() 
```

```python
# Replace missing value in the dataframe

# finding mean ca column
meanCa = np.mean(df.Ca)

# fill the missing values (na => not available values) with the mean value
df.Ca = df.Ca.fillna(meanCa)

df.isna().sum()
```

```python
# To remove missing values
df = df.dropna()

# Now we'll see count of missing value in each column
df.isna().sum()
```
